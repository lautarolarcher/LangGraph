```python
## Ejercicio 1
# Crear un agente simple, con input: {name: Lautaro} y output: Lautaro, estudia burro, sino la IA te va a dejar sin laburo
# Concatenar el estado no reemplazarlo
```

```python
from typing import Dict, TypedDict
from langgraph.graph import StateGraph
```

```python
class AgentState (TypedDict):
    name: str
```

```python

def agent_saludo(state:AgentState) -> AgentState:
    """Esta agente genera una advertencia al nombre ingresado"""
    state["name"] =  "Hola " + state["name"] + " ponete a estudiar burro, la IA te va a sacar el laburo"
    return state
```

```python

graph = StateGraph(AgentState)
graph.add_node("saludo", agent_saludo)
graph.set_entry_point("saludo")
graph.set_finish_point("saludo")
app=graph.compile()
```

```python
from IPython.display import Image, display
display(Image(app.get_graph().draw_mermaid_png()))
```

```python


result=app.invoke({"name":"Lautaro"})
```
