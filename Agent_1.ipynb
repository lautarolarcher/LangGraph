#### Hello World Agent

```python
from typing import Dict ,TypedDict, Optional, Any
# estas son Type Annotations
from langgraph.graph import StateGraph
```

##### State Graph

```python
#lo primero es definir el estado del grafo
class Agentstate(TypedDict): #este es el schema state
    messagge: str
```

#### Node

```python
#primer nodo
#este es el agente que tiene como input el estado y como output el estado
def hello_world_agent(state:Agentstate) -> Agentstate: 
    """Es un agente simple que recibe un nombre y devuelve un saludo"""
    state["messagge"] = "Hello " + state["messagge"] + " how is going?" #actualiza el estado
    return state #devuelve el estado para pasarlo al siguiente nodo
```

#### Grafico

```python
from langgraph.graph import StateGraph

graph = StateGraph(Agentstate)
graph.add_node("hello_wordl_agent", hello_world_agent)
#aca hay que definir el nombre y la accion que realiza el nodo
graph.set_entry_point("hello_wordl_agent")#defino que nodo es el que se conecta con el inicio
graph.set_finish_point("hello_wordl_agent")#defino que nodo es el que se conecta con el fin

app =graph.compile()
```

```python
from IPython.display import Image, display
display(Image(app.get_graph().draw_mermaid_png()))
```

```python

result = app.invoke({"messagge": "Bob"})
```
